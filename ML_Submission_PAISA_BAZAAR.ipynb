## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Describing all the columns of the data

print(f"Columns in the data are : {data.columns}")

print(f"Numerical Columns are : {numerical_cols}")
print(f"Number of numerical columns : {len(numerical_cols)}")

print(f"Categorical Columns are : {categorical_cols}")
print(f"Number of categorical columns : {len(categorical_cols)}")
```

```python
# Data Wrangling

# Drop duplicate rows based on ID
data = data.drop_duplicates(subset=['ID'])

# Standardize categorical formatting
data['Credit_Mix'] = data['Credit_Mix'].str.title()  # e.g., 'good' → 'Good'
data['Payment_of_Min_Amount'] = data['Payment_of_Min_Amount'].replace({'NM': 'No'})

# Convert Credit_History_Age (months) to years
data['Credit_History_Years'] = (data['Credit_History_Age'] / 12).round(1)

# Create Debt-to-Income ratio
data['Debt_to_Income'] = (data['Outstanding_Debt'] / (data['Annual_Income'] + 1)).round(3)

# Dropping Out Unnecessary columns
df = data.drop(['ID', 'Name', 'SSN'],axis=1)
print(f"Shape of the new dataset is : {df.shape}")
```

### What all manipulations have you done and insights you found?

* Data Manipulations Performed

    1.Removed duplicate records

    Dropped any duplicate rows based on the unique identifier ID to ensure each record represents a unique observation.

    2.Standardized categorical formatting

    Converted inconsistent text cases in Credit_Mix (e.g., good → Good).

    Replaced NM with No in Payment_of_Min_Amount for clarity and uniformity.

    3.Feature Engineering

    Credit_History_Years: Created a new column by converting Credit_History_Age (in months) into years, making it easier to interpret.

    Debt_to_Income: Added a derived metric representing the ratio of Outstanding_Debt to Annual_Income, giving insights into customer debt burden.

    4.Dropped unnecessary columns

    Removed ID, Name, and SSN since they are identifiers and carry no analytical value.

* Insights / Result

    > The dataset was cleaned, standardized, and enriched with new analytical features.

    > No missing values were present, so no imputation was required.

    > Final shape of the dataset: (100000, 27)
    > This indicates that after dropping 3 identifier columns and adding 2 engineered features, the final dataset is ready for analysis or modeling.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

> Null hypothesis (H0): There is no difference in annual income between customers with good and poor credit scores.

> Alternate hypothesis (H1): Customers with a good credit score have a higher annual income than customers with poor credit scores.

#### 2. Perform an appropriate statistical test.

```python
import scipy.stats as stats
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
from scipy.stats import shapiro

# categorize
good_income = df[df['Credit_Score'] == 'Good']['Annual_Income']
poor_income = df[df['Credit_Score'] == 'Poor']['Annual_Income']
```

```python
# Checking normality
print(shapiro(good_income))
print(shapiro(poor_income))
```

```python
# Plotting the variable to check the distribution
plt.figure(figsize=(12, 10))

# 1️⃣ Histogram for Good Credit
plt.subplot(2, 2, 1)
sns.histplot(good_income, kde=True, color='teal')
plt.title("Good Credit Score - Annual Income Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")

# 2️⃣ Histogram for Poor Credit
plt.subplot(2, 2, 2)
sns.histplot(poor_income, kde=True, color='coral')
plt.title("Poor Credit Score - Annual Income Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")

# 3️⃣ Q–Q Plot for Good Credit
plt.subplot(2, 2, 3)
stats.probplot(good_income, dist="norm", plot=plt)
plt.title("Q–Q Plot: Good Credit Score")

# 4️⃣ Q–Q Plot for Poor Credit
plt.subplot(2, 2, 4)
stats.probplot(poor_income, dist="norm", plot=plt)
plt.title("Q–Q Plot: Poor Credit Score")

plt.tight_layout()
plt.show()
```

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import mannwhitneyu

u_stat, p_value = mannwhitneyu(good_income, poor_income, alternative='greater')

print("U-statistic:", u_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject H₀: Customers with good credit scores have significantly higher annual income.")
else:
    print("Fail to reject H₀: No significant difference in income.")
```

##### Which statistical test have you done to obtain P-Value?

> Mann–Whitney U Test

##### Why did you choose the specific statistical test?

> Because normality is violated and sample size is large, use the Mann–Whitney U test (non-parametric alternative to t-test).

```python
# Detect and Count Outliers
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    return len(outliers), lower, upper

for col in num_cols:
    n_outliers, lower, upper = detect_outliers_iqr(df, col)
    print(f"{col:<30} | Outliers: {n_outliers}")
```

```python
# selectively cap only the skewed columns
cols_to_cap = [
    'Annual_Income', 'Monthly_Inhand_Salary', 'Delay_from_due_date',
    'Changed_Credit_Limit', 'Num_Credit_Inquiries', 'Outstanding_Debt',
    'Total_EMI_per_month', 'Amount_invested_monthly', 'Monthly_Balance', 'Debt_to_Income'
]

for col in cols_to_cap:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.clip(df[col], lower, upper)
```

```python
for col in cols_to_cap:
    n_outliers, _, _ = detect_outliers_iqr(df, col)
    print(f"{col:<30} | Outliers after capping: {n_outliers}")
```

### 1. Which Evaluation metrics did you consider for a positive business impact and why?

| Metric                                         | Why it Matters for Business Impact                                                                                                                                                                     |
| ---------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| **Accuracy**                                   | Provides overall correctness of the model. Helps gauge general reliability but can be misleading if classes are imbalanced.                                                                            |
| **Precision (per class)**                      | Measures how many predicted positives are actually correct. For credit scoring, high precision in identifying **good credit customers** avoids granting loans to risky individuals, reducing defaults. |
| **Recall (per class)**                         | Measures how many actual positives are correctly identified. High recall for **risky/poor credit customers** ensures potential defaulters are detected, reducing financial losses.                     |
| **F1-Score**                                   | Balances precision and recall. Critical when a trade-off exists between catching risky customers and avoiding false positives, giving a holistic view of model performance.                            |
| **Class-wise metrics**                         | Especially important for multi-class credit scores (Good, Average, Poor) to ensure **all risk levels are properly detected**, not just the majority class.                                             |
| **Weighted metrics**                           | Accounts for **class imbalance**, ensuring that performance on minority classes (like high-risk customers) is not ignored.                                                                             |
| **ROC-AUC / PR-AUC (if using binary version)** | Helps understand model discrimination ability, especially for identifying high-risk vs low-risk customers. Higher AUC → better separation between risky and safe borrowers.                            |


> Business Rationale:

    Minimizing defaults: High recall for poor credit customers avoids lending to risky individuals.

    Maximizing opportunity: High precision for good credit customers ensures profitable lending.

    Balanced decisions: F1-score and weighted metrics prevent the company from focusing only on the majority group, ensuring fair and profitable credit decisions.
